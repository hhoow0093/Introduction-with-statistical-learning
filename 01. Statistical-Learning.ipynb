### np.random.normal() -> generate random normal data
*parameters:*
1. *loc* - determines the mean of the gennerated random data
2. *scale* - determines the standard deviation from the random generated data
3. *size* - how many data is being generated

- if a standard devation data is **increasing**, most of the spread of data are getting further away from the mean
- if a stamdard deviation data is **decreasing**, most of the spread of the data are getting closer with the mean

```python
x = np.random.normal(size=50)
x
```

```python
y = x + np.random.normal(loc=50, scale=1, size=50)
```

### np.corrcoef(x, y)

uses pearson correlation method to calculate relationship between 2 variables.
- (0, 0) - shows correlation between x and x
- (0, 1) - shows correlarion between x and y
- (1, 0) - shows correlation between y and x
- (1, 1) - shows correlation between y and y

```python
np.corrcoef(x, y)
```

## Graphics

```python
import numpy as np
from matplotlib.pyplot import subplots

# Random data
rng = np.random.default_rng()
x = rng.standard_normal(100)
y = rng.standard_normal(100)

# Create 3 subplots stacked horizontally
fig, axes = subplots(1, 3, figsize=(18, 5) )

# 1️⃣ First plot - red plus markers
axes[0].plot(x, y, 'r+')
axes[0].set_title("Plot 1: 'r+' (red plus markers)")

# 2️⃣ Second plot - blue circle markers
axes[1].plot(x, y, 'bo')
axes[1].set_title("Plot 2: 'bo' (blue circles)")

# 3️⃣ Third plot - default line plot
axes[2].plot(x, y)
axes[2].set_title("Plot 3: default line")

# Add labels and layout
for ax in axes:
    ax.set_xlabel("x")
    ax.set_ylabel("y")

fig.tight_layout()

```

## For loops

```python
total = 0
for value in [3,2,19]:
  total += value
  print('Total is: {0}'.format(total))
```

```python
total = 0
for value in [2,3,19]:
  for weight in [3, 2, 1]:
    total += value * weight
    print('Total is: {0}'.format(total))
```

```python
total = 0
for value, weight in zip([2,3,19],[0.2,0.3,0.5]):
  total += weight * value
  print('Weighted average is: {0}'.format(total))
```

## String formatting

```python
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5))
M = rng.choice([0, np.nan], p=[0.8,0.2], size=A.shape)
A += M
D = pd.DataFrame(A, columns=['food',
                            'bar',
                            'pickle',
                            'snack',
                            'popcorn'])
D[:3]
```

```python
for col in D.columns:
  template = 'Column "{0}" has {1:.2%} missing values'
  print(template.format(col,
  np.isnan(D[col]).mean()))
```
